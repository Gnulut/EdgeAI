```python
import pandas as pd

# Sample data
data = {'A': [1, 1, 2, 2, 2, 3, 3],
        'B': ['X', 'X', 'Y', 'Y', 'Z', 'Z', 'Z']}
df = pd.DataFrame(data)

# Detect changes in 'A' or 'B'
df['group'] = (df['A'] != df['A'].shift()) | (df['B'] != df['B'].shift())
df['group'] = df['group'].cumsum()  # Create unique group numbers

# Print groups
for _, group in df.groupby('group'):
    print(group, '\n')
    print(_)
```
