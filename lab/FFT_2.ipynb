- 時域信號的 y 軸應標示為 "Amplitude"（振幅）
- 頻域信號的 y 軸應標示為 "Magnitude"（幅值）

```python
import numpy as np
import matplotlib.pyplot as plt

# 假設 wesab._df['ECG'][:100] 是一個時域信號
# 這裡生成一個模擬的 ECG 信號作為示例
np.random.seed(0)
time_signal = np.sin(2 * np.pi * 5 * np.linspace(0, 1, 100)) + 0.5 * np.random.randn(100)


# 計算 FFT
freq_signal = np.fft.fft(time_signal)
freqs = np.fft.fftfreq(len(time_signal), d=1/100)  # 假設取樣率為 100 Hz

# 只取正頻率部分
half_idx = len(time_signal) // 2
half_freqs = freqs[:half_idx]
half_magnitude = np.abs(freq_signal[:half_idx])

magnitude = np.abs(freq_signal)

# 繪製時域和頻域圖
fig, ax = plt.subplots(3, 1, figsize=(10, 6))

# 時域信號
ax[0].plot(time_signal, label="ECG Signal")
ax[0].set_title("Time Domain Signal")
ax[0].set_xlabel("t")
ax[0].set_ylabel("Amplitude")
ax[0].legend()

# 頻域信號
ax[1].stem(half_freqs, half_magnitude, 'b',basefmt=" ")
ax[1].set_title("Frequency Domain Signal (FFT)")
ax[1].set_xlabel("Frequency (Hz)")
ax[1].set_ylabel("Magnitude")

ax[2].stem(freqs, magnitude, 'g',basefmt=" ")
plt.tight_layout()
plt.show()
```

```python
phase = np.angle(freq_signal)  # 計算相位
plt.plot(freqs, phase)
plt.title("Phase Spectrum")
plt.xlabel("Frequent (Hz)")
plt.ylabel("Phase (radians)")
plt.show()
```
